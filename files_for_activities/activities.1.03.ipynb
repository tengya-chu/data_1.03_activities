```python
cols = [] # lista vacía
for i in range(len(data.columns)): # primera vez entrara al 0, luego al 1, luego al 2, etc. 
    cols.append(data.columns[i].lower()) # cuando la i valga 0, me ira a la columna 0, con el append quiere decir que se vaya añadiendo en la lista cols
```

```python
cols #hemos creado una lista con los nombres en minuscula
```

```python
data.columns = cols # para cambiar las columnas del DataFrame 'data' a los items que hay en la lista 'cols'
```
